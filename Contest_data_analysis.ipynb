## 1. Which shop have I spent the most money in February

## 2. Show a logic to find the shop with highest expenditure in each type (use rank window function)

```python
expen_df
```

```python
hig_exp = expen_df.groupby(['type','shop'])['cost'].sum()
```

```python
hig_exp
```

```python
expen_df['rank'] = expen_df.groupby('type')['cost'].rank(ascending=False, method='dense')
```

```python
highest_expenditure_shops_by_type = expen_df[expen_df['rank'] == 1][['type', 'shop', 'cost']]
```

```python
print(highest_expenditure_shops_by_type)
```

## 3. Calculate the % increase in expenses from 2nd February to 3rd February.

```python
expen_df
```

```python
cost_2_feb = expen_df[expen_df['date'] == '2/2/2024']['cost'].sum()
cost_2_feb
```

```python
cost_3_feb = expen_df[expen_df['date'] == '2/3/2024']['cost'].sum()
cost_3_feb
```

```python
percentage_increase = ((cost_3_feb - cost_2_feb) / cost_2_feb) * 100
```

```python
print('The percentage Increase in expenses from 2nd Feb to 3rd Feb is', percentage_increase,'%')
```
